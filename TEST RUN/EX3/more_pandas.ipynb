```python
dataset1 = pd.read_excel('../../Lesson04/Exercise48-51/Sample - Superstore.xls',
                         sheet_name='Missing')
dataset1.head(5)
```

```python
for column in dataset1.columns:
    nulls = dataset1[column].isnull().sum()
    print(f'{column} has {nulls} nulls')
```

```python
dataset1[['Customer', 'Product']].fillna('Unknown')
```

```python
dataset1[['Sales', 'Quantity', 'Profit']].fillna(0)
```

```python
dataset1[['Sales']].fillna(method='bfill')
```

```python
dataset1[['Quantity']].fillna(method='ffill')
```

```python
dataset1[['Profit']].fillna(dataset1.mean()['Profit'])
```

```python
dataset1.dropna(axis=0).reset_index()
```

```python
dataset1.dropna(axis=1, thresh=10)
```
